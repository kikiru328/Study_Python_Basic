```python
import pandas as pd
```

## Load Dataset

```python
train = pd.read_csv("titanic/train.csv")
train.head()
```

### Explore

```python
# matplotlib 시각화 툴. 하지만 잘 안씀.
# matplotlib은 앞에 % 를 쓰는게 특징
# Seaborn 이 보편화 되어있음.
%matplotlib inline
import seaborn as sns
```

### Countplot

```python
sns.countplot(data=train, x="Embarked")
```

```python
sns.countplot(data=train, x="Embarked", hue="Survived")
```

```python
sns.countplot(data=train, x="Pclass")
```

```python
sns.countplot(data=train, x="Pclass", hue="Survived")
```

### barplot

```python
sns.barplot(data=train, x="Pclass", y="Fare")
```

* Barplot은 y축을 지정함. (숫자 데이터만 가능)
* 검은색 바 - 신뢰구간. (검은색 바를 넘어가면 Outlier)

```python
sns.barplot(data=train, x="Pclass",y="Fare",hue="Survived")
```

* 엑셀에서 볼 수 없는 것을 데이터 시각화로 판단할 수 있음
* 연속적인 데이터는 barplot. 

## Pointplot

```python
sns.pointplot(data=train, x="Pclass",y="Fare",hue="Survived")
# Pointplot은 시간데이터등을 볼 수 있음. (시간별 매출현황 등)
```

* bar와 point=> X축 차이. X축 관에 연관성 없음 = Bar. 있으면 = Point

## lmplot

```python
# LMPLOT (L소문자)
# 아래선 = 회귀선 -> 잘 안쓰임 (reg.   없애려면 fit_reg=False)
sns.lmplot(data=train, x="Age", y="Fare")
```

```python
sns.lmplot(data=train, x="Age", y ="Fare", hue="Survived", fit_reg=False)
```

* 회귀선 삭제 (fit_reg=False)
* 데이터들에서 동떨어진 데이터 = Out-lier

```python
# Out-lier 삭제 (Fare <500)
low_fare = train[train["Fare"]<500]
sns.lmplot(data=low_fare, x="Age", y="Fare", hue="Survived", fit_reg=False)
```

```python
# 100단위 미만
low_fare=train[train["Fare"]<100]
sns.lmplot(data=low_fare, x="Age", y="Fare", hue="Survived", fit_reg=False)
```

* lmplot은 연속적 데이터가 두개를 활용할 때 좋음

## distplot

* 쓰는 방법이 조금 다름
* 데이터를 찾아오고 distplot()으로 감싸줌 
* dist => distribution (분포도)

```python
sns.distplot(train["Fare"])
```

```python
sns.distplot(low_fare["Fare"])
```

* Gausisian = 정규분포. 가장 이상적인 distplot .
* 나온다고 하더라도 의심해야함. 이미 정리되었나 등등

## Visualization _ Seaborn

```python
import seaborn as sns
```

```python
# 출력은 화면저장 혹은 바로 띄우기

%matplotlib inline

# 퍼센티지 꼭 써주기
```

## countplot

x축은 알아서데이타값으로

```python
sns.countplot(data=train, x = 'Embarked', hue = 'Survived')

# hue에 해당된 변수로 데이터를 나누는 것 따로 시각화, 비교할 수 있겠!
```

```python
sns.countplot(data=train, x = 'Pclass', hue = 'Survived')
```

```python
sns.countplot(data=train, x = 'Pclass')
```

## barplot

```python
# x, y 데이터를 직접 지정해줘야하며 y축 값은 숫자만 --> 평균값으로 적용됨
```

```python
sns.barplot(data=train, x='Pclass', y="Fare")
```

```python
# 파란색 바는 신뢰구간을 뜻함 검은바 안에 있으면 non-outliar, 밖에 있으면 outliar
# 전반적인 분포에서 동떨어진 것들을 outliar
```

```python
sns.barplot(data=train, x='Pclass', y="Fare", hue = 'Survived')
```

```python
# 연속적 데이터는 barplot(age, fare) y축에다가 연속적데이터
```

## pointplot

## lmplot

```python
sns.lmplot(data=train, x = "Age", y = "Fare")
```

```python
sns.lmplot(data=train, x = "Age", y = "Fare", hue = 'Survived', fit_reg=False)



# 회귀선 그리지 않고, Survived랑 비교
```

```python
# outliar 눈에 확 들어옴
# outliar 제거필요 
```

```python
# outliar제거!

low_fare = train[train['Fare']<500]

sns.lmplot(data=low_fare, x = "Age", y = "Fare", hue = 'Survived', 
           fit_reg=False)
```

```python
# 100달러 포커싱

low_fare = train[train['Fare']< 100]

sns.lmplot(data=low_fare, x = "Age", y = "Fare", hue = 'Survived',
           
           fit_reg=False)
```

```python
# lmplot은 연속적인 데이터가 2개일 때 사용 (x, y데이터 모두 연속성데이터일경우)
```

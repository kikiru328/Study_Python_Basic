```python
train = pd.read_csv("train.csv", index_col="PassengerId")

print(train.shape)
train.head()
```

##  Explore

## Preprocessing

### Encoding Sex

```python
# 전처리 -> 코딩해줘야함 

train.head(2)
```

```python
train['Sex']
# 바꾸기

train['Sex_encoded'] = \
train['Sex'].replace("female", 1).replace("male", 0)

# train file 에 Sex를 코딩해줌

cols = ["Sex", "Sex_encoded"]
train[cols]

#보고싶은 것만 꺼내보기
# 위 처럼 하면은 sex 와 sex_encoded와 함께 나와서 정확한 판단 가능


train[cols].head()
```

```python
test['Sex']

test['Sex_encoded'] = \
test['Sex'].replace("female", 1).replace("male", 0)

cols = ["Sex", "Sex_encoded"]
test[cols]

print(test.shape)
# 행렬 수 확인할 때 사용


test[cols].head()

# !!! train column의 Sex를 코딩함
```

```python
train["Pclass"].head(5)

# pclass 확인하기 -> 이미 숫자로 되어있으니 별도의 코딩 필요 없음
```

### Fill in Fare

```python
train['Fare']

# fare 은 숫자여서 코딩 별도로 안해도 괜찮, but NaN 확인해야함
```

```python
train[train['Fare'].isnull()]

# isnull= NaN 찾을 때 사용
```

```python
test[test['Fare'].isnull()]
```

```python
# 채워주기 fillna
# 항상 검색부터 하기

test['Fare'].fillna(0)

# NaN 있으면 fillna기능으로 0 처리 << 0 처리하는건 자유
```

```python
# 덮어씌우기

test["Fare"] = test['Fare'].fillna(0)
```

```python
test[test['Fare'].isnull()]

# 덮어씌워졌나 확인하고 
# 원본파일 건드는거
```

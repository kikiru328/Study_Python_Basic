### weather_bar plot EDA

```python
sns.barplot(data=train, x='weather', y='count')
```

```python
figure, ((ax1, ax2, ax3)) = plt.subplots(ncols=3)
figure.set_size_inches(18,8)

sns.barplot(data=train, x='sunny', y='count', ax=ax1)
sns.barplot(data=train, x='cloudy', y='count', ax=ax2)
sns.barplot(data=train, x='awful', y='count', ax=ax3)
```

### windspeed

```python
train['windspeed'].value_counts()
```

```python
plt.figure(figsize = (18, 4))

sns.distplot(train['windspeed'])

sns.distplot(data = train, x='windspeed', y='count')
```

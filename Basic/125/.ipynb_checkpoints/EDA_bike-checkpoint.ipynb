## 데이터준비

### Datetime

```python
train["datetime"]
```

```python
train["datetime"] = pd.to_datetime(train["datetime"])
```

```python
train["datetime-year"] = train["datetime"].dt.year
train["datetime-month"] = train["datetime"].dt.month
train["datetime-day"] = train["datetime"].dt.day
train["datetime-hour"] = train["datetime"].dt.hour
train["datetime-minute"] = train["datetime"].dt.minute
train["datetime-second"] = train["datetime"].dt.second

print(train.shape)
train.head()
```

## weather encoding

```python
train['weather_encoded'] = \
train['weather'].replace(1, 'sunny').replace(2, 'cloudy'). replace(3, 'awful')
```

```python
train.head(50)
```

```python
data = train['weather_encoded']
```

```python
weather_dummies = pd.get_dummies(data)
```

```python
weather_dummies
```

```python
train.head(1)
```

```python
train['sunny'] = weather_dummies['sunny']
```

```python
train['cloudy'] = weather_dummies['cloudy']
```

```python
train['awful'] = weather_dummies['awful']
```

```python
train.head()
```

```python
print(train.shape)
train.head(1)
```

```python
train.to_csv("train-1.csv", index=False)
```

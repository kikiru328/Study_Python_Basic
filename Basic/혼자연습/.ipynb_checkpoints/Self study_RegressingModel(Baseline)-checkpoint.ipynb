## Preprocessing

```python
# data를 컬럼으로 전처리하기
# pandas의 dateime기술 이요
# 파일명["datetime-year"] = 파일명['datetime'].dt.year
```

```python
pd.to_datetime(train["datetime"])

train['datetime'] = pd.to_datetime(train["datetime"])

# 이거 두개는 항상 꼭 가져와야댐!
# 항상 test도 똑같이
```

```python
train['datetime-year'] = train['datetime'].dt.year
train['datetime-month'] = train['datetime'].dt.month
train['datetime-day'] = train['datetime'].dt.day
train['datetime-hour'] = train['datetime'].dt.hour
train['datetime-minute'] = train['datetime'].dt.minute
train['datetime-second'] = train['datetime'].dt.second

print(train.shape)
train.head()


# 이렇게해줬으면 test도 똑같이 해줘야함
```

```python
pd.to_datetime(test["datetime"])

test['datetime'] = pd.to_datetime(test["datetime"])

# 항상 train하고 test도 똑같이
```

```python
test['datetime-year'] = test['datetime'].dt.year
test['datetime-month'] = test['datetime'].dt.month
test['datetime-day'] = test['datetime'].dt.day
test['datetime-hour'] = test['datetime'].dt.hour
test['datetime-minute'] = test['datetime'].dt.minute
test['datetime-second'] = test['datetime'].dt.second

print(test.shape)
test.head()

```

## X, Y 축 설정하기

```python
feature_name = ['datetime-year', 'datetime-month', 'datetime-day', 
                'datetime-hour', 'datetime-minute', 'datetime-second', 'season', 'holiday','workingday', 
               'weather', 'temp', 'atemp','humidity', 'windspeed']
```

```python
feature_name
```

```python
# feature지정해 준 후 다음에 y 값 label지정 (구하고 싶은 것)

label_name = 'count'
label_name
```

```python
X_train = train[feature_name]

#선언하고 항상 확인할 것

print(X_train.shape)
X_train.head()
```

```python
y_train = train[label_name]

print(y_train.shape)
y_train.head()
```

```python
X_test = test[feature_name]
```

```python
X_test.head()
```

## Preprocessing

### Encode Sex

```python
train["Sex_encode"] = train["Sex"].replace("male", 0).replace("female", 1)

train[["Sex", "Sex_encode"]].head()
```

```python
test["Sex_encode"] = test["Sex"].replace("male", 0).replace("female", 1)

test[["Sex", "Sex_encode"]].head()
```

### Fill in missing Fare

```python
# NaN(Not a Number), null
test.loc[test["Fare"].isnull(), "Fare"] = 0

test[test["Fare"].isnull()]
```

### Group Age

```python
train["Child"] = train["Age"] < 10

train[["Age", "Child"]].head()
```

```python
test["Child"] = test["Age"] < 10

test[["Age", "Child"]].head()
```

### Train

```python
# Feature(X) - Pclass, Sex, Fare, ...
# Label(y) - Survived
```

```python
feature_names = ["Pclass", "Sex_encode", "Fare", "Child"]
feature_names
```

```python
X = train[feature_names]

X.head()
```

```python
label_name = "Survived"

y = train[label_name]

y.head()
```

### Use Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

# 1) fit -> 2) predict

model = DecisionTreeClassifier(max_depth = 5)
model
```

```python
model.fit(X, y)
```

### Visualize

```python
import graphviz
from sklearn.tree import export_graphviz

tree = export_graphviz(model,
                       feature_names = feature_names,
                       class_names = ["Perish", "Survived"])

graphviz.Source(tree)
```

### Predict

```python
X_test = test[feature_names]

X_test.head()
```

```python
prediction = model.predict(X_test)
prediction
```

### Submit

```python
submit = pd.read_csv("data/gender_submission.csv")

submit["Survived"] = prediction

submit.head()
```

```python
submit.to_csv("decision-tree.csv", index = False)
```

### pivot table

```python
# 특정 조건을 걸고 그것에 대한 기계통계치

order
pd.pivot_table(order, index = 'state',values = 'price')

# 구분인덱스에 값에 맞게 분류하고 계산필요value의 값을(mean, sum)을 분석해줌
```

```python
pd.pivot_table(order, 
               index = 'state',
               values = 'price',
              aggfunc = 'sum')
```

```python
pd.pivot_table(order, 
               index = 'state',
               values = 'price',
              aggfunc = ['mean', 'sum'])
```

```python
pd.pivot_table(order, 
               index = 'state',
               values = 'price',
              aggfunc = [np.mean, np.sum])
```

```python
pd.pivot_table(order, 
               index = ['date','state'],
               values = 'price',
              aggfunc = 'sum')
```

```python
pd.pivot_table(order, 
               index = ['state','date'],
               values = 'price',
              aggfunc = 'sum')
```

### apply

```python
def is_vip(price):
    if price >= 500:
        return "VIP"
    else:
        return "Non-VIP"
    
order['level'] = order['price'].apply(is_vip)
order


# 컬럼명과 인자명이 같을 필요는 없음
```

```python
def is_vip(row):
    if row['state'] == 'confirmed' and row['price'] >= 500:
        return "VIP"
    else:
        return "Non-VIP"
    
order['level'] = order.apply(is_vip, axis = "columns")
order

# 데이터프레임 전체에 apply하려면 axis = columns 무조건
```

### 여러개의 데이터를 하나로 합치기

```python
url = "http://bit.ly/dsa-02-order-201701"

order01 = pd.read_csv(url)
order01
```

```python
url = "http://bit.ly/dsa-02-order-201702"

order02 = pd.read_csv(url)
order02
```

```python
order = pd.concat([order01, order02])

order.sort_values (by = 'user_id')
```

```python
user = pd.read_csv('user.csv')
user
```

```python
product = pd.read_csv('product.csv')
product
```

```python
pd.merge(order, user, left_on = 'user_id', right_on = 'id')

# 왼쪽에 있는 user_id와 오른쪽의 id와 합쳐라
```

```python
user['user_id'] = user['id']

pd.merge(order, user, on = 'user_id')
```

```python
# 데이터 없는 사람들은 버리자 (교집합만) - inner
# 데이터 없는 사람들 살리자 Nan값 (합집합) - outer

user['user_id'] = user['id']

pd.merge(order, user, on = 'user_id', how = 'inner')
```

```python
pd.merge(order, user, on = 'user_id', how = 'outer')
```

```python
# 왼쪽만 살리고 오른쪽 버리고 - left
# 오른쪽만 살리고 왼쪽은 버리고 - right

pd.merge(order, user, on = 'user_id', how = 'left')
```

```python
pd.merge(order, user, on = 'user_id', how = 'right')
```

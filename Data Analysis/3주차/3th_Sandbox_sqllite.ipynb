```python
query = """
    SELECT 
        date,
        price,
        state
    FROM orders
    WHERE state <> 'confirmed'
    AND price >= 500
    
"""

order = pd.read_sql(query, connection)
order
```

```python
query = """
    SELECT 
        date,
        price,
        state
    FROM orders
    WHERE price
    BETWEEN 500 AND 900
    
"""

order = pd.read_sql(query, connection)
order
```

```python
query = """
    SELECT 
        *
    FROM orders
    WHERE address IN ('Seoul', 'Incheon','Busan')
    
"""

order = pd.read_sql(query, connection)
order
```

```python
query = """
    SELECT
        orders.date,
        orders.price,
        orders.state,
        users.name,
        users.gender,
        users.age
    FROM orders
        LEFT JOIN users
            ON orders.user_id = users.id

"""

order = pd.read_sql(query, connection)
order
```
